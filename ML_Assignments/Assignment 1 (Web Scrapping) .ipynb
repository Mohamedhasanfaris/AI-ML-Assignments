```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
url="https://www.flipkart.com/search?q=asus+Gaming+laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&p%5B%5D=facets.rating%255B%255D%3D4%25E2%2598%2585%2B%2526%2Babove"
```

```python
response=requests.get(url)
response
```

```python
html=BeautifulSoup(response.text)
html
```

```python
Name=html.select("._4rR01T")
Name
```

```python
Rating=html.select("._3LWZlK")
Rating
```

```python
Price=html.select("._1_WHN1")
Price
```

```python
Processor=html.select(".rgWa7D:nth-child(1)")
Processor
```

```python
Ram=html.select(".rgWa7D:nth-child(2)")
Ram
```

```python
Display=html.select(".rgWa7D:nth-child(5)")
Display
```

```python
Storage=html.select(".rgWa7D:nth-child(4)")
Storage
```

```python
Details=pd.DataFrame(Name,columns=["Gaming Laptop Name"])
Details["Processor"]=Processor
Details["Ram"]=Ram
Details["Display"]=Display
Details["Storage"]=Storage
Details["Rating"]=Rating
Details["Price"]=Price

Details
```
